```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(x,y1,color='green',label="lichi")
plt.plot(x,y2,color='yellow',label="mango")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of fruit over time  
plt.legend()
plt.show()
```

###Bar Graph

```python
x_cordinates=np.array([0,1,2])*2
plt.style.use("dark_background")
plt.bar(x_cordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"])
plt.bar(x_cordinates+0.25,[20,10,12],width=0.5,label="Next Year")
plt.legend()
plt.title("Price")
plt.ylim(0,30)
plt.xlim(-0.5,5)
plt.show()
```
